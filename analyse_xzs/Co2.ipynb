```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

数据来源：
二氧化碳排放量 世界银行网站


数据处理：
利用pandas读入csv文件并进行预处理
删除无用字段

```python
co2 = pd.read_csv("/Users/wowblk/Desktop/数据集/温室气体排放数据/API_EN.ATM.CO2E.KT_DS2_zh_csv_v2_5361673.csv")
co2 = co2.drop('Indicator Name',axis=1)
co2 = co2.drop('Indicator Code',axis=1)
co2
```

```python
co2[co2['Country Name'] == '中国']
```

```python
co2[co2['Country Name'] == '美国']
```

```python
co2_China = co2[co2['Country Name'] == '中国'].iloc[0,32:66]
co2_China
```

```python
co2_America = co2[co2['Country Name'] == '美国'].iloc[0,32:66]
co2_America
```

```python
data = pd.DataFrame({'year': co2_China.index,
                     'co2_China': co2_China.values,
                     'co2_America': co2_America.values
                     })
fig, ax = plt.subplots(figsize=(8,6))
sns.lineplot(x='year', y='value', hue='variable', data=pd.melt(data, ['year']), ax=ax)
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.set_xlabel('Year')
ax.set_ylabel('Volume')
ax.legend(fontsize=10, facecolor='white', edgecolor='black', fancybox=True, shadow=True, loc='center left', bbox_to_anchor=(1, 0.05))
plt.grid(alpha=0)
ax.set_xlim(left=0)
sns.despine()

plt.show()
```
