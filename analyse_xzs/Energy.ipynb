```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import matplotlib
matplotlib.rc("font",family='SimHei')
```

```python
energy = pd.read_csv("D:\WorkSpace\Code\Python\EnergyTrans\dataset\energy\Energy_Transition.csv")
energy
```

```python
energy = energy.drop('CTS_Code',axis=1)
energy = energy.drop('CTS_Name',axis=1)
energy = energy.drop('CTS_Full_Descriptor',axis=1)
energy = energy.drop('ISO2',axis=1)
energy = energy.drop('ISO3',axis=1)
energy.columns = energy.columns.str.replace('F','')
energy = energy.drop('Source',axis=1)
energy
```

```python
df = energy[energy['Country/Region'] == 'China, P.R.: Mainland']
df
```

```python

Bioenergy = energy[energy['Country/Region'] == 'China, P.R.: Mainland'].iloc[0,6:27]
Geothermal_energy = energy[energy['Country/Region'] == 'China, P.R.: Mainland'].iloc[4,6:27]
Hydropower = energy[energy['Country/Region'] == 'China, P.R.: Mainland'].iloc[6,6:27]
Marine_energy = energy[energy['Country/Region'] == 'China, P.R.: Mainland'].iloc[8,6:27]
Solar_energy = energy[energy['Country/Region'] == 'China, P.R.: Mainland'].iloc[16,6:27]
Wind_energy = energy[energy['Country/Region'] == 'China, P.R.: Mainland'].iloc[18,6:27]

data = pd.DataFrame({'year': Bioenergy.index,
                     '生物能': Bioenergy.values,
                     '地热能': Geothermal_energy.values,
                     '水电能': Hydropower.values,
                     '海洋能': Marine_energy.values,
                     '太阳能': Solar_energy.values,
                     '风能': Wind_energy.values})

fig, ax = plt.subplots(figsize=(8,6))
sns.lineplot(x='year', y='value', hue='variable', data=pd.melt(data, ['year']), ax=ax)
plt.yscale('log')
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.set_xlabel('年份')
ax.set_ylabel('发电量')
ax.set_xlim(left=0)

ax.legend(fontsize=10, facecolor='white', edgecolor='black', fancybox=True, shadow=True, loc='center left', bbox_to_anchor=(1, 0.15))
plt.grid(alpha=0)
sns.despine()

plt.show()
```

```python
import seaborn as sns

sns.set_style("whitegrid")
sns.set_palette("husl")

# Extract data for China
china_energy = energy[energy['Country/Region'] == 'China, P.R.: Mainland']

# Define x-axis data as years
x = china_energy.columns[6:].astype(int)

# Extract y-axis data for each energy type
y1 = china_energy.iloc[0, 6:].values
y2 = china_energy.iloc[4, 6:].values
y3 = china_energy.iloc[6, 6:].values
y4 = china_energy.iloc[8, 6:].values
y5 = china_energy.iloc[16, 6:].values
y6 = china_energy.iloc[18, 6:].values

# Create plot with subplots
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(16, 8), sharex=True, sharey=True)

# Define plot titles
titles = ['生物能', 'Geothermal energy', 'Hydropower', 'Marine energy', 'Solar energy', 'Wind energy']

# Define plot colors
colors = sns.color_palette("husl", 6)

# Plot each energy type in its own subplot
for i, ax in enumerate(axs.flat):
    ax.plot(x, eval(f'y{i+1}'), color=colors[i], label=titles[i])
    ax.set_title(titles[i], fontsize=14)
    ax.set_ylabel('Volume (log scale)', fontsize=12)
    ax.set_yscale('log')
    ax.yaxis.set_tick_params(labelsize=10)
    ax.xaxis.set_tick_params(labelsize=10)
    ax.grid(alpha=0.3)
    ax.spines["top"].set_alpha(0.0)
    ax.spines["bottom"].set_alpha(0.3)
    ax.spines["right"].set_alpha(0.0)
    ax.spines["left"].set_alpha(0.3)
    ax.legend(fontsize=10, loc='best')
    ax.grid(alpha=0)

# Add x-axis label to bottom subplots only
axs[1, 1].set_xlabel('Year', fontsize=14)

ax.set_xlim(left=2000)

# Add overall plot title
plt.suptitle('Renewable Energy Consumption in China', fontsize=18)

# Adjust subplot spacing
plt.subplots_adjust(hspace=0.4, wspace=0.25)

# Show plot
plt.show()
```

```python
Total_renewable_energy = Bioenergy + Geothermal_energy + Hydropower + Marine_energy + Solar_energy + Wind_energy
Total_renewable_energy
```

```python
from matplotlib import pyplot as plt
#新增加的两行
import matplotlib
matplotlib.rc("font",family='SimHei')
import seaborn as sns



x = Total_energy.columns
y1 = Total_energy.sum().values.astype(float)
y2 = Total_renewable_energy.values.astype(float)



ratio = y2 / y1# 标注峰值点数值



fig, ax = plt.subplots(figsize=(8,6))




sns.lineplot(x=x, y=ratio, ax=ax, color='#2196F3', linewidth=2.5)
plt.fill_between(x, y1=0,y2=ratio,alpha=0.2, color='#2196F3')

ax.set_title('可再生能源消耗量占比', fontsize=22, fontweight='bold', pad=20)
ax.set_xlabel('年份', fontsize=16, labelpad=15)
ax.set_ylabel('可再生能源消耗量占比', fontsize=16, labelpad=15)
ax.set_xticks(x)
ax.set_xticklabels(x, fontsize=14)
ax.set_yticks([0.1, 0.2, 0.3])
ax.set_xlim(x[0],x[-1])

ax.set_yticklabels(['10%', '20%', '30%'], fontsize=14)
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.spines['top'].set_alpha(0)
ax.spines['bottom'].set_alpha(1)
ax.spines['right'].set_alpha(0)
ax.spines['left'].set_alpha(1)
ax.set_ylim([0, 0.3])


plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
#新增加的两行
import matplotlib
matplotlib.rc("font",family='SimHei')
matplotlib.rcParams['axes.unicode_minus']=False

Bio_rate = (Bioenergy[20] / Total_renewable_energy[20]) * 100
Geo_rate = (Geothermal_energy[20] / Total_renewable_energy[20]) * 100
Hydro_rate = (Hydropower[20] / Total_renewable_energy[20]) * 100
Marine_rate = (Marine_energy[20] / Total_renewable_energy[20]) * 100
Solar_rate = (Solar_energy[20] / Total_renewable_energy[20]) * 100
Wind_rate = (Wind_energy[20] / Total_renewable_energy[20]) * 100

data = [Bio_rate,Hydro_rate,Solar_rate,Wind_rate]
labels = ['生物能','水利能','太阳能','风能']
fig, ax = plt.subplots(figsize=(8, 6))
colors = ['#007acc', '#00ba00', '#ff7f0e', '#ab003c', '#ffa500']

wedges, texts, autotexts = ax.pie(data, labels=None, startangle=80, pctdistance=0.85,
                                  wedgeprops={'linewidth': 2, 'edgecolor': 'white', 'linestyle': 'solid'},
                                  autopct='%1.1f%%', textprops={'fontsize': 12}, colors=colors,
                                   labeldistance=None, shadow=True,)

plt.setp(autotexts, size=12, weight='bold', color='white')



plt.title('清洁能源分类占比图', fontsize=18, fontweight='bold', y=1.1)

# Set legend properties
legend_properties = {'family': 'serif', 'size': 14, 'weight': 'bold', 'style': 'italic', 'edgecolor': 'black', 'facecolor': 'white'}

# 设置图例
ax.legend(labels, loc='upper left', bbox_to_anchor=(0.9, 1),
          prop={'size': 12, 'weight': 'bold'},
          edgecolor='black', frameon=True)
plt.tight_layout()
plt.show()
```
