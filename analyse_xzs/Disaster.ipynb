```python
import matplotlib.pyplot as plt

labels = ['Hydrological', 'Meteorological', 'Biological', 'Geophysical', 'Climatological']
data1 = disaster[disaster['Disaster Subgroup'] == 'Hydrological']['Disaster Type'].value_counts().sum()
data2 = disaster[disaster['Disaster Subgroup'] == 'Meteorological']['Disaster Type'].value_counts().sum()
data3 = disaster[disaster['Disaster Subgroup'] == 'Biological']['Disaster Type'].value_counts().sum()
data4 = disaster[disaster['Disaster Subgroup'] == 'Geophysical']['Disaster Type'].value_counts().sum()
data5 = disaster[disaster['Disaster Subgroup'] == 'Climatological']['Disaster Type'].value_counts().sum()

data = [data1, data2, data3, data4, data5]

fig, ax = plt.subplots(figsize=(8, 6))
explode = (0, 0.1, 0, 0, 0.1)
colors = ['#007acc', '#00ba00', '#ff7f0e', '#ab003c', '#ffa500']

wedges, texts, autotexts = ax.pie(data, labels=None, startangle=80, pctdistance=0.85,
                                  wedgeprops={'linewidth': 2, 'edgecolor': 'white', 'linestyle': 'solid'},
                                  autopct='%1.1f%%', textprops={'fontsize': 12}, colors=colors,
                                  explode=explode, labeldistance=None, shadow=True)

plt.setp(autotexts, size=12, weight='bold', color='white', fontfamily='serif')



plt.title('Disaster', fontsize=18, fontweight='bold', fontfamily='serif', y=1.1)

# Set legend properties
legend_properties = {'family': 'serif', 'size': 14, 'weight': 'bold', 'style': 'italic', 'edgecolor': 'black', 'facecolor': 'white'}

# 设置图例
ax.legend(labels, loc='upper left', bbox_to_anchor=(0.9, 1),
          prop={'family': 'serif', 'size': 12, 'weight': 'bold'},
          edgecolor='black', frameon=True)

plt.tight_layout()
plt.show()
```
